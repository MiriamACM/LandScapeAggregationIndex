```python
import geopandas as gpd
from shapely.geometry import polygon
import matplotlib.pyplot as plt
```

```python
# Catalonia map for fuel 1
```

```python
fuel1 = gpd.read_file("/home/coca/Maps_TFM/Polygon/Europe/fuelCategory1.geojson")
```

```python
fuel1['geometry'].crs
```

```python
fuel1_proj=fuel1.to_crs(25831)
```

```python
fuel1_proj['geometry'].crs
```

```python
cat_shape = gpd.read_file("/home/coca/Maps_TFM/Maps_Cat_ICGC/divisions-administratives-v2r1-catalunya-1000000-20220801.shp")
```

```python
cat_shape['geometry'].crs
```

```python
CatFuel1 = fuel1_proj.overlay(cat_shape, how ='intersection')
```

```python
CatFuel1.plot()
```

```python
CatFuel1.to_file("/home/coca/Maps_TFM/Polygon/Catalonia/CatFuel1.geojson",driver='GeoJSON')
```
