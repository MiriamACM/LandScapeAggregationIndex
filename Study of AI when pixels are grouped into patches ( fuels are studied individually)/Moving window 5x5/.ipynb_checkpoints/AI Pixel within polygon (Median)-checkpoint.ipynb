```python
import pandas as pd
import numpy as np
import rasterio
import geopandas as gpd
from shapely.geometry import Point
from rasterio.crs import CRS

import matplotlib.pyplot as plt
```

```python
# This is the code to calculate the median of aggregation index of the pixels that are within polygons.
```

```python
#This is the code to get the coordinates of the raster's pixels in a dataframe

def pixels2points (dataset, polygon):
    val = dataset.read(1) # band 1
    geometry = [Point(dataset.xy(x,y)[0],dataset.xy(x,y)[1]) for x,y in np.ndindex(val.shape) if val[x,y] != -999999]
    v = [val[x,y] for x,y in np.ndindex(val.shape) if val[x,y] != -999999]
    df = gpd.GeoDataFrame({'geometry':geometry,'data':v})
    df.crs = polygon.crs
    return df
```

```python
# This is a code to calculate the median of the AI and the Area of the pixels that are within a specific polygon

def AI_pixels_within_polygon(df_match):
    
    AI_median_list = []
    Area_list = []
    Poly_ID_array = df_match['ID'].unique() # This is an array with all polygons ID that contain points
    Patch_ID_list = []
    
    
    for ii in Poly_ID_array:
       
        part = df_match.loc[df_match['ID'] ==ii]
        median = part['data'].median()
        AI_median_list.append (median)
        Area_list.append(len(part))
        Patch_ID_list.append(ii)
    
    
    df= pd.DataFrame()
    df['AI (patch)']=AI_median_list
    df['Area_Ha (patch)'] = Area_list
    df['ID (patch)'] = Patch_ID_list
   
    
    return  df
```

```python
# This is a code to calculate the category of the AI. 
#Use this function with the dataframe in which there is the info about the pixels and the polygons

def category_calculation (df,text1, text2):

    fuel_list = []
    cat_0_list = []
    cat_1_list = []
    cat_2_list = []
    
    if text1 == 'patch':
        cat_2 = df[df['AI (patch)'] ==0 ].count()[0]
        cat_1 = df[(0< df['AI (patch)']) & (df['AI (patch)']<=0.90)].count()[0]
        cat_0 = df[df['AI (patch)'] >0.90 ].count()[0]
        total = cat_2 + cat_1 + cat_0
        
        cat_2_list.append(cat_2/total*100)
        cat_1_list.append(cat_1/total*100)
        cat_0_list.append(cat_0/total*100)
    
    if text1 == 'polygon':
        cat_2 = df[df['AI'] ==0 ].count()[0]
        cat_1 = df[(0< df['AI']) & (df['AI']<=0.90)].count()[0]
        cat_0 = df[df['AI'] >0.90 ].count()[0]
        total = cat_2 + cat_1 + cat_0
        
        cat_2_list.append(cat_2/total*100)
        cat_1_list.append(cat_1/total*100)
        cat_0_list.append(cat_0/total*100)
    
    fuel_list.append(text2)
    
    df_new = pd.DataFrame()
    df_new['Fuel'] = fuel_list
    df_new['Category 0'] = cat_0_list
    df_new['Category 1'] = cat_1_list
    df_new['Category 2'] = cat_2_list
    
    return df_new
```

```python
# This code is to calculate the ratio between the Area of the polygon and the related patch's Area

def ratio (df):
    

    df['Ratio (%)'] = df.apply(lambda x: (((x['Area_Ha']) - x['Area_Ha (patch)'])/(x['Area_Ha']))*100, axis=1)

    
    return df
```

```python
# Get the points that are within the polygons
CatFuel1_RinP = gpd.tools.sjoin(CatFuel1R_point,CatFuel1P, predicate = 'within', how = 'inner')
CatFuel2_RinP = gpd.tools.sjoin(CatFuel2R_point,CatFuel2P, predicate = 'within', how = 'inner')
CatFuel3_RinP = gpd.tools.sjoin(CatFuel3R_point,CatFuel3P, predicate = 'within', how = 'inner')
CatFuel4_RinP = gpd.tools.sjoin(CatFuel4R_point,CatFuel4P, predicate = 'within', how = 'inner')
CatFuel5_RinP = gpd.tools.sjoin(CatFuel5R_point,CatFuel5P, predicate = 'within', how = 'inner')
```

```python
# Apply the function to calculate the median of the AI of the pixels that are within the polygons

Median_AI_Fuel1 = AI_pixels_within_polygon(CatFuel1_RinP)
Median_AI_Fuel2 = AI_pixels_within_polygon(CatFuel2_RinP)
Median_AI_Fuel3 = AI_pixels_within_polygon(CatFuel3_RinP)
Median_AI_Fuel4 = AI_pixels_within_polygon(CatFuel4_RinP)
Median_AI_Fuel5 = AI_pixels_within_polygon(CatFuel5_RinP)
```

```python
# Join the Median_AI_FuelX to the CatFuelX to see the Area and the median of the AI for the pixels that are within the polygons

AI_patch_poly_Fuel1 = CatFuel1P.merge(Median_AI_Fuel1, how='inner',  left_on='ID',right_on='ID (patch)').drop(['ID (patch)'],axis=1)
AI_patch_poly_Fuel2 = CatFuel2P.merge(Median_AI_Fuel2, how='inner',  left_on='ID',right_on='ID (patch)').drop(['ID (patch)'],axis=1)
AI_patch_poly_Fuel3 = CatFuel3P.merge(Median_AI_Fuel3, how='inner',  left_on='ID',right_on='ID (patch)').drop(['ID (patch)'],axis=1)
AI_patch_poly_Fuel4 = CatFuel4P.merge(Median_AI_Fuel4, how='inner',  left_on='ID',right_on='ID (patch)').drop(['ID (patch)'],axis=1)
AI_patch_poly_Fuel5 = CatFuel5P.merge(Median_AI_Fuel5, how='inner',  left_on='ID',right_on='ID (patch)').drop(['ID (patch)'],axis=1)
```

```python
# Modification of the Area of polygons
AI_patch_poly_Fuel1.loc[:,'Area_Ha'] = AI_patch_poly_Fuel1.area
AI_patch_poly_Fuel2.loc[:,'Area_Ha'] = AI_patch_poly_Fuel2.area
AI_patch_poly_Fuel3.loc[:,'Area_Ha'] = AI_patch_poly_Fuel3.area
AI_patch_poly_Fuel4.loc[:,'Area_Ha'] = AI_patch_poly_Fuel4.area
AI_patch_poly_Fuel5.loc[:,'Area_Ha'] = AI_patch_poly_Fuel5.area
```
