```python
import geopandas as gpd
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
def poly_loghist(x, bins):
    hist, bins = np.histogram(x, bins=bins)
    logbins = np.logspace(np.log10(bins[0]),np.log10(bins[-1]),len(bins))
    plt.hist(x, bins=logbins,color ='cadetblue', edgecolor = 'black')
    plt.xscale('log')
    plt.yscale('log')
    plt.xlim(10**(-2),10**7)
    plt.ylim(0,10**5)
    plt.xlabel('Area (Ha)')
    plt.ylabel('Number of polygons')
    plt.grid()
```

```python
# Map of Catalonia with Fuel 1 (Polygon format)
EUFuel1P= gpd.read_file("/home/coca/Maps_TFM/Polygon/Europe/fuelCategory1.geojson")
```

```python
# Modification of the Area of polygons
EUFuel1P.loc[:,'Area_Ha'] = EUFuel1P.area/10000
```

```python
# Number of patches Vs. Area // Number of polygons Vs. Area
# Using lognormal distribution in the X axis 

fig, ax = plt.subplots(nrows = 1, ncols= 1, figsize=(10,5))

plt.subplot(1,2,2)
poly_loghist(EUFuel1P['Area_Ha'], 100)

plt.title ('Area distribution of Fuel Category 1 (Polygon)')

plt.tight_layout()
plt.show()

#plt.savefig("/home/coca/Maps_TFM/Raster/Output/Fuel1/AI_Dist_Fuel1.jpg")
```
